```python
population = pd.read_csv('2021_population.csv')
population.head()
```

```python
population.drop(['2020_population', 'area','density_sq_km', 'growth_rate', 'world_%', 'rank'], axis=1, inplace=True) 
```

```python
population.rename(columns={'2021_last_updated':'population'}, inplace=True)
```

```python
population.info()
```

```python
population['population'] = population['population'].str.replace(',', '')
```

```python
population['population'] = population['population'].astype(int)
```

```python
population.info()
```

```python
data_pop = pd.merge(data, population, how='left', on='country')
```

```python
data_pop.info()
```

```python
data_pop = data_pop.assign(positivity_rate = lambda x: x['confirmed'] / x['population'])
```

```python
data_pop = data_pop.assign(mortality_rate = lambda x: x['deaths'] / x['confirmed'])
```

```python
#data_pop = data_pop.assign(total_active = lambda x: x['confirmed'] - x['recovered'] - x['deaths'])
```

```python
# Active Case = confirmed - deaths - recovered
data_pop['Active'] = data_pop['confirmed'] - data_pop['deaths'] - data_pop['recovered']
```

```python
data_pop.head()
```

```python
data_pop.to_csv('data_pop.csv')
```

```python
full_grouped = data_pop.groupby(['date', 'country'])['confirmed', 'deaths', 'recovered', 'Active'].sum().reset_index()
```

```python
full_grouped.head()
```

```python
# new cases 
temp = full_grouped.groupby(['country', 'date', ])['confirmed', 'deaths', 'recovered']
temp = temp.sum().diff().reset_index()
mask = temp['country'] != temp['country'].shift(1)
temp.loc[mask, 'confirmed'] = np.nan
temp.loc[mask, 'deaths'] = np.nan
temp.loc[mask, 'recovered'] = np.nan
# renaming columns
temp.columns = ['country', 'date', 'New cases', 'New deaths', 'New recovered']
# merging new values
full_grouped = pd.merge(full_grouped, temp, on=['country', 'date'])
# filling na with 0
full_grouped = full_grouped.fillna(0)
# fixing data types
cols = ['New cases', 'New deaths', 'New recovered']
full_grouped[cols] = full_grouped[cols].astype('int')
# 
full_grouped['New cases'] = full_grouped['New cases'].apply(lambda x: 0 if x<0 else x)
```

```python
full_grouped.head()
```

```python
full_grouped.to_csv('full.csv')
```
