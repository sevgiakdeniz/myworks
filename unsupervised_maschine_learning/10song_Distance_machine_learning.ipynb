```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./df_audio_features.csv') 
```

```python
df.set_index(['song_name', 'artist'], inplace=True)
```

```python
all_distances = []
for song_name in df.index:

    danceability = df.loc[song_name][0]
    energy = df.loc[song_name][1]
    key = df.loc[song_name][2]
    loudness = df.loc[song_name][3]
    mode = df.loc[song_name][4]
    speechiness = df.loc[song_name][5]
    acousticness = df.loc[song_name][6]
    instrumentalness = df.loc[song_name][7]
    liveness = df.loc[song_name][8]
    valence = df.loc[song_name][9]
    tempo = df.loc[song_name][10]
    duration_ms = df.loc[song_name][11]
    time_signature = df.loc[song_name][12]
    
    song_name_distance = []
    
    for song_name_to_compare in df.index:
        
       
        danceability_to_compare= df.loc[song_name_to_compare][0]
        energy_to_compare = df.loc[song_name_to_compare][1]
        key_to_compare = df.loc[song_name_to_compare][2]
        loudness_to_compare = df.loc[song_name_to_compare][3]
        mode_to_compare = df.loc[song_name_to_compare][4]
        speechiness_to_compare = df.loc[song_name_to_compare][5]
        acousticness_to_compare = df.loc[song_name_to_compare][6]
        instrumentalness_to_compare = df.loc[song_name_to_compare][7]
        liveness_to_compare = df.loc[song_name_to_compare][8]
        valence_to_compare = df.loc[song_name_to_compare][9]
        tempo_to_compare = df.loc[song_name_to_compare][10]
        duration_ms_to_compare = df.loc[song_name_to_compare][11]
        time_signature_to_compare = df.loc[song_name_to_compare][12]
        
       
        danceability_distance = abs(danceability - danceability_to_compare)
        energy_distance = abs(energy - energy_to_compare)
        key_distance = abs(key - key_to_compare)
        loudness_distance = abs(loudness - loudness_to_compare)
        mode_distance = abs(mode - mode_to_compare)
        speechiness_distance = abs(speechiness - speechiness_to_compare)
        acousticness_distance = abs(acousticness - acousticness_to_compare)
        instrumentalness_distance = abs(instrumentalness - instrumentalness_to_compare)
        liveness_distance = abs(liveness - liveness_to_compare)
        valence_distance = abs(valence - valence_to_compare)
        tempo_distance = abs(tempo - tempo_to_compare)
        duration_ms_distance = abs(duration_ms - duration_ms_to_compare)
        time_signature_distance = abs(time_signature - time_signature_to_compare)
        
        distance = danceability_distance + energy_distance + key_distance + loudness_distance + mode_distance + speechiness_distance + acousticness_distance + instrumentalness_distance + liveness_distance + valence_distance + tempo_distance + duration_ms_distance + time_signature_distance
        
        print(f'The distance between {song_name} and {song_name_to_compare} is {distance}')
        
        song_name_distance.append(distance)
        
    print(f'The distances from {song_name} are {song_name_distance}')
        
    all_distances.append(song_name_distance)
        
```

```python
all_distances
```

```python
song_name_manhattan_distance = pd.DataFrame(all_distances)
```

```python
song_name_manhattan_distance.columns = df.index
```

```python
import seaborn as sns
plt.subplots(figsize=(12,8))
sns.heatmap(song_name_manhattan_distance)
```

```python
# can you do the same process than before for all the columns?
# create a function to do it for all the users
from math import sqrt
all_distances = []

for song_name in df.index: 

    ratings = [rating for rating in df.loc[song_name]]
    song_name_distances = []
    
    for song_name_to_compare in df.index: 
        ratings_to_compare = [rating_to_compare for rating_to_compare in df.loc[song_name_to_compare]]
        # euclidean distance formula
        distance = sqrt(sum([(ratings[pos] - ratings_to_compare[pos])**2 for pos in range(len(ratings))]))
        song_name_distances.append(distance)
    
    all_distances.append(song_name_distances)

st_euclidean_distances = pd.DataFrame(all_distances, columns =  df.index, index =  df.index)
plt.subplots(figsize=(12, 8))
sns.heatmap(st_euclidean_distances)
```

```python
def min_max_scaling(col):
    min_col = min(col)
    max_col = max(col)
    norm_val = []
    for val in col:
        norm_val.append((val - min_col) / (max_col - min_col))
    return norm_val
```

```python
pd.DataFrame({
    'original': df.iloc[:,0],
    'min_max_scaling': min_max_scaling(df.iloc[:,0])
})
```
